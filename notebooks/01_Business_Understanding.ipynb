Credit card fraud is one of the most significant financial crimes worldwide, costing billions of dollars annually. This project builds an end-to-end machine learning pipeline to **automatically detect fraudulent credit card transactions** in real time.

The dataset contains transactions made by European cardholders over **two days in September 2013**:

| Property | Detail |
|---|---|
| Total Transactions | 284,807 |
| Fraudulent Transactions | 492 (~0.17%) |
| Features | 30 (V1–V28 via PCA, + Time, Amount) |
| Target | `Class` — 0 = Legitimate, 1 = Fraud |

All features except `Time` and `Amount` have been **PCA-transformed** to protect user confidentiality.

## 2. Problem Statement

Financial institutions process **millions of transactions per day**. Manually reviewing each transaction for fraud is infeasible. The key challenges are:

- **Extreme class imbalance** — Fraudulent transactions represent only **~0.17%** of all transactions. A naive model predicting "all legitimate" achieves 99.83% accuracy but is completely useless.
- **Evolving fraud patterns** — Fraudsters constantly change tactics, making static rule-based systems ineffective.
- **Cost asymmetry** — The cost of missing a fraud (false negative) far exceeds the cost of a false alarm (false positive).
- **Speed requirement** — Fraud decisions must be made in **milliseconds** at transaction time.

> **Core Question:** *Can we build a model that reliably identifies the rare fraudulent transactions while minimising inconvenience to legitimate customers?*

## 3. Business Objective

The primary business objectives are:

1. **Maximise fraud detection (Recall)** — Catch as many fraudulent transactions as possible to minimise financial losses.
2. **Minimise false positives (Precision)** — Avoid blocking legitimate transactions, which damages customer trust and experience.
3. **Balance the trade-off** — Use metrics like **F1-score**, **PR-AUC**, and **ROC-AUC** that account for imbalance.
4. **Provide explainability** — Model decisions must be interpretable for compliance and audit purposes.

### Priority Ordering
```
Recall > Precision > Latency > Interpretability
```
A missed fraud costs the bank real money. A blocked legitimate transaction costs customer goodwill.

## 4. Stakeholders

| Stakeholder | Role | Key Concern |

| **End Customers** | Cardholders | Seamless experience, account security |

## 5. Business Impact

### Financial Impact
- The global cost of credit card fraud exceeded **$33 billion in 2022** (Nilson Report).
- Each undetected fraud transaction results in direct financial loss to the issuing bank.
- A **1% improvement in recall** on a bank processing 10M transactions/month with avg fraud value of $150 ≈ **$150,000+ in prevented losses**.

### Reputational Impact
- Customers who experience fraud and feel it was not caught quickly **switch banks**.
- Excessive false positives (wrongly blocked cards) also drive churn.

### Operational Impact
- Automating detection **reduces manual review workload** by fraud analysts.
- Enables real-time blocking of fraudulent transactions before settlement.

## 6. Machine Learning Formulation

### Task Type
**Binary Classification** — predict whether a transaction is fraudulent (`Class = 1`) or legitimate (`Class = 0`).

### Input Features
- `V1` – `V28`: PCA-transformed anonymised features
- `Time`: Seconds elapsed since the first transaction in the dataset
- `Amount`: Transaction amount in euros

### Target
- `Class`: `0` = Legitimate, `1` = Fraud

### Key Modelling Decisions
| Decision | Choice | Rationale |
|---|---|---|
| Imbalance handling | SMOTE, class weights, undersampling | Fraud is only 0.17% of data |
| Primary metric | **PR-AUC / F1** | ROC-AUC is misleading on imbalanced data |
| Model family | Ensemble (XGBoost, LightGBM, RF) | High performance on tabular data |
| Threshold | Tuned post-training | Business cost trade-off |
| Explainability | SHAP values | Required for compliance |

## 7. Success Criteria

The model will be considered successful if it meets the following thresholds on the **held-out test set**:

| Metric | Minimum Target | Stretch Goal |
|---|---|---|
| **Recall (Fraud)** | ≥ 0.80 | ≥ 0.90 |
| **Precision (Fraud)** | ≥ 0.70 | ≥ 0.85 |
| **F1-Score (Fraud)** | ≥ 0.75 | ≥ 0.87 |
| **PR-AUC** | ≥ 0.80 | ≥ 0.90 |
| **ROC-AUC** | ≥ 0.95 | ≥ 0.98 |

> **Note:** Recall is weighted more heavily than Precision. Missing a fraud costs more than a false alarm.

## 8. Assumptions & Limitations

### Assumptions
- The PCA-transformed features (`V1`–`V28`) retain sufficient information for fraud detection even without interpretability.
- The two-day snapshot is representative of broader transaction patterns.
- The dataset labels are accurate and were verified by domain experts.

### Limitations
| Limitation | Impact |
|---|---|
| **Anonymised features** | Cannot perform feature engineering with domain knowledge |
| **Static dataset** | Real-world fraud patterns drift over time — model will need retraining |
| **No merchant / geography data** | Important fraud signals are missing |
| **Two-day window** | Seasonal patterns and long-term behaviour not captured |
| **Single currency (EUR)** | May not generalise to other markets |
