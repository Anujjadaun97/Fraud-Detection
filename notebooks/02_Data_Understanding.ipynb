## 7. Missing Value Analysis

Check if there are any missing (NaN/Null) values in the dataset that would require imputation.

```python
missing_values = df.isnull().sum()
missing_cols = missing_values[missing_values > 0]

if missing_cols.empty:
    print("No missing values found in the dataset.")
else:
    print("Missing values found:\n", missing_cols)
```

```python
# Missing Percentage
missing_percentage = (missing_values / len(df)) * 100

missing_percentage
```

## 8. Duplicate Analysis

Identify if there are any perfectly identical rows in the dataset. While some duplicates might be legitimate in transactional data, it's important to quantify them.

```python
duplicate_count = df.duplicated().sum()
duplicate_pct = (duplicate_count / len(df)) * 100

print(f"Number of duplicate rows: {duplicate_count:,}")
print(f"Percentage of duplicates: {duplicate_pct:.4f}%")
```

## 9. Target Variable Analysis

Examine the distribution of the `Class` column to quantify the class imbalance.

```python
class_counts = df['Class'].value_counts()
class_pcts = df['Class'].value_counts(normalize=True) * 100

target_dist = pd.DataFrame({
    'Count': class_counts,
    'Percentage (%)': class_pcts
})

target_dist.index = ['Legitimate (0)', 'Fraudulent (1)']
target_dist
```

## 11. Initial Data Quality Assessment

Based on the checks above, we assess the overall quality of the dataset:

- **Anomalies:** Scale differences (Time/Amount vs PCA components)

## 12. Key Observations

1. **Imbalance:** The dataset is heavily imbalanced, with frauds accounting for only ~0.17%.
2. **Scaling:** The PCA components (`V1`–`V28`) are already scaled/transformed. However, `Time` and `Amount` are in their original scales and will need normalization/standardization before feeding into distance-based models or neural networks.
3. **Completeness:** There are no missing values in this dataset.
4. **Duplicates:** There is a small percentage of duplicate rows. We will need to decide whether to drop them in the preprocessing stage.
